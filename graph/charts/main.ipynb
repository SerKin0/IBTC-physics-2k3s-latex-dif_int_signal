```python
%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "CMU Serif",
    "font.size": 16
})
```

```python
C = 10
R = 5
```

```python
def K(w):
    return 1 / np.sqrt(1 + (w * C * R)**2)
def Phi(w):
    return np.arctan(-w * C * R)
```

```python
w = np.linspace(0, 0.1, 1000)
x = np.linspace(0.0001, 0.1, 1000)
y_k = [K(w_) for w_ in w]
y_w = [1 / (w_ * C * R) for w_ in x]

plt.ylim(0, 1.5)
plt.xlim(0, 0.1)

plt.plot([0, 0.2], [1, 1], linewidth='0.8', color='black')
plt.plot(w, y_w, linestyle='dashed')
plt.plot(x, y_k)

plt.yticks([0, 1], ['0', '1'])
plt.xticks([0],[0])
plt.xlabel('$w$')
plt.ylabel('$K$')

plt.savefig('1.pdf', dpi=300, transparent=True)
```

```python
w = np.linspace(0, 1, 1000)

y_p = [Phi(x_) for x_ in w]
plt.plot(w, y_p)
plt.plot([0, 1], [-np.pi/2, -np.pi/2], linestyle='dashed')
plt.plot([0, 0], [-3, 1], linewidth='0.8', color='black')
plt.plot([-1, 3], [0, 0], linewidth='0.8', color='black')

plt.ylim(-1.65, 0.1)
plt.xlim(-0.01, 0.6)

plt.yticks([0, -np.pi/2], ['0', r'$-\frac{\pi}{2}$'])
plt.xticks([0], [0])
plt.xlabel('$w$')
plt.ylabel('$arg(\dot K)$')

plt.savefig('2.pdf', dpi=300, transparent=True)
```

```python
def K(w):
    return 1 / np.sqrt(1 + 1 / (w * C * R)**2)
def Phi(w):
    return np.arctan(1 / (w * C * R))
```

```python
w = np.linspace(0, 0.2, 1000)

y_k = [K(w_) for w_ in w]
y_w = [w_ * C * R for w_ in w]

plt.plot(w, y_w, linestyle='dashed')
plt.plot(w, y_k)
plt.plot([0, 0.2], [1, 1], linewidth='0.8', color='black')

plt.xlim(0, 0.1)
plt.ylim(0, 1.5)

plt.yticks([0, 1], ['0', '1'])
plt.xticks([0], [0])
plt.xlabel('$w$')
plt.ylabel('$K$')

plt.savefig('3.pdf', dpi=300, transparent=True)
```

```python
w = np.linspace(0, 0.2, 1000)

y_p = [Phi(x_) for x_ in w]
plt.plot(w, y_p)
plt.plot([0, 0.2], [np.pi/2, np.pi/2], linestyle='dashed')
plt.plot([-0.1, 0.2], [0, 0], linewidth=0.8, color='black')
plt.plot([0, 0], [-0.1, 2], linewidth=0.8, color='black')

plt.ylim(-0.07, 2)
plt.xlim(-0.005, 0.2)

plt.yticks([0, np.pi/2], ['0', r'$\frac{\pi}{2}$'])
plt.xticks([0], [0])
plt.xlabel('$w$')
plt.ylabel('$arg(\dot K)$')

plt.savefig('4.pdf', dpi=300, transparent=True)
```
